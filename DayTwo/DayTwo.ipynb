**Part One**

```python
with open("DayTwo.txt", "r") as f:
    txt_list = [str(line.strip()) for line in f]
txt_list

rows = []
chechSumList = []
for i in txt_list: 
    rowsInt = [int(i) for i in i.split("\t")]
    rows.append(rowsInt)
    chechSumList.append(max(rowsInt) - min(rowsInt))
print(sum(chechSumList))
```

**Part Two**

```python
divRowsList = []
for i in rows:
    for j in i:
        for k in i:
          if(k==j):
            continue
          else:
            if(k%j == 0):
              divRowsList.append(k // j)
print(sum(divRowsList))
```
